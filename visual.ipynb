```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_excel("cleaned_dataset.xlsx")
df
```

<h1>Plots</h1>

1. unvariate plot

Q1.How many employees are there in each status category?

```python
sns.countplot(data=df,x='status')
```

Observation - There are more Active employees present compared than Terminated employees

```python
sns.countplot(data=df,x='gender',hue="status")
# hue is used for color based grouping
```

Q2.How many Active and Terminated employees are there in each experience level?

```python
sns.countplot(data=df,x="exp_level",hue="status")
```

Observation - 
1. There are approximately the same amount of active and terminated junior employees(Terminated are slightly more)
2. There are more active  mid-senior employees than terminated mid-senior employees with a big gap
3. There are more acctive senior employees than terminated senior employees with a even larger gap

```python
plt.figure(figsize=(5,2)) #(width,height)
sns.countplot(data=df,x="exp_level",hue="status")
plt.title('Experience Level')
plt.show() # It displays data while also removing unneccesary 
```

<h1>Pie Chart</h1>

Q3.Draw a pie chart from the attrition dataset that shows the proportion of different termination types among employees

```python
df['term_type'].value_counts
val = df['term_type'].value_counts().values
idx = df['term_type'].value_counts().index
print(val,idx)
plt.figure(figsize=(3,3))
plt.pie(data=df,x=val,labels=idx,autopct='%.1f%%')
plt.title('Termination Type')
plt.show()
```

Observation - 
1. Minimal Involuntary termination
2. Mediocre Voluntary termination
3. Massive number of Active employees  

<h1>Kernel Density Estimation (KDE)</h1>

Q4.plot a smooth curve to understand how employee ages are distributed

```python
plt.figure(figsize=(5,2))
plt.title('Age Distribution using KDE plot')
sns.kdeplot(data=df,x="age")
plt.grid()
plt.xticks(range(10,80,5))
plt.show()
```

Observation - Plot is left skewed indicating lesser employees between age of 20 to 60. Larger concentration on the right for ages 60 to 65

Q5.Create a histogram to visualize the distribution of employee ages


```python
plt.figure(figsize=(5,2))
sns.histplot(data=df,x="age")
plt.title("Age Distribution using Hist plot")
plt.xticks(range(20,75,5))
plt.show()
```

Observation - More senior employees(in the range of 60 and 65)

<h1>Bivariate</h1>

Barplot

Q1.Show the average experience of employees for each gender

```python
plt.figure(figsize=(5,2))
sns.barplot(data=df,x="gender",y="experience",estimator='mean')
plt.title("Gender vise average experience")
plt.show()
```

Observation - Both male and female employees have an average experience of more than 10 years

Q2.Show the maximum experience of employees for each gender and status

```python
plt.figure(figsize=(8,4))
sns.barplot(data=df,x="gender",y="experience",estimator='max',hue='status')
plt.title("Gender vise Highest experience")
plt.legend(loc="upper right", edgecolor="black",fontsize=7,borderpad=1)
plt.show()
```

Observation - In both Male and Female and in both active and terminated employees the highest experience is 25 and more

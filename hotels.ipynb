```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('new_york_hotels.csv',encoding='latin-1')
df
```

```python
df.info()
```

```python
# sns.boxplot(df['star_rating'])
```

Observation - No outlier so using mean

```python
mean_star = df['star_rating'].mean().round(1)
mean_star
```

```python
df['star_rating'] = df['star_rating'].fillna(mean_star)
df.info()
```

```python
df.duplicated().any()
```

Observation - No duplicate rows

```python
df['state_province'].unique()
df = df.drop('state_province',axis=1)
```

```python
df = df.rename({'ean_hotel_id':'hotel_id'},axis=1)
df
```

```python
df['fare'] = df['high_rate']-df['low_rate']
df
```

Count hotels per city

```python
df.groupby('city').agg({'city':'count'})
```

Calculate how many hotels have each star_rating

```python
df['star_rating'].value_counts()
```

For each star_rating, compute mean, median, min, and max of high_rate and low_rate.



```python
df.groupby('star_rating').agg({'high_rate':['mean','median','min','max'],'low_rate':['mean','median','min','max']})
```

Hotels with “0.0” rates

```python
# df.query('star_rating==0.0')
df.loc[df['star_rating']==0.0,['hotel_id','city']]
```

Fare range per hotel

```python
df = df.sort_values('fare',ascending=True)
df = df.reset_index(drop=True)
df.head()
```

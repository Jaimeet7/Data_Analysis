1. create a dataframe called product (id,aname,price) which includes 5 products
2. create a jan_sales and feb_sales (id,sold) which includes 3 and 4 sales respectively
3. concat jan_sales and feb_sales as sales_df 
4. create a series of loc and concat with sales_df
5. merge the product and sales 

```python
d = {'id':['p1','p2','p3','p4','p5'],
    'name':['pen','book','charts','markers','white board'],
    'price':[10,50,5,30,800]}

product = pd.DataFrame(d)
```

```python
sales_1 = {'id':['p1','p2','p2'],'sold':[20,10,30]}
jan_sales = pd.DataFrame(sales_1)

sales_2 = {'id':['p4','p2','p1','p5'],'sold':[100,5,70,1]}
feb_sales = pd.DataFrame(sales_2)
```

```python
sales = pd.concat([jan_sales,feb_sales],ignore_index=True) # Default axis=0
```

```python
location = pd.Series(['Banglore','Mumbai','Chennai','Hyderabad','Uttarakhand'],name='loc')
sales = pd.concat([sales,location],axis=1)
```

```python
pd.merge(left=product,right=sales,on='id',how='outer')
# how can take('inner','outer','left outer','right outer','cross'). 
```

1. create a student_df with (name,stream,py,sql,web) attributes and 5 values
2. perform melt operation
3. perform pivot operation 
4. perform pivot_table

```python
d2 = {
    'name':['Jaimeet','Rahul','Nishant','Pratham','Sindhu'],
    'stream':['CSE','ECE','CSE','ISE','ME'],
    'py':[100,98,99,97,96],
    'sql':[80,85,90,80,85],
    'web':[85,70,90,80,75]
}
student = pd.DataFrame(d2)
student
```

```python
student.melt(id_vars=['name','stream'],value_vars=['py','sql','web'],var_name='subjects',value_name='marks')
```

```python
melt_df = pd.melt(student,id_vars=['name','stream'],value_vars=['py','sql','web'],var_name='subjects',value_name='marks')
# melt -> column to rows (unpivoting)
melt_df
```

```python
pd.pivot(melt_df,index=['name','stream'],columns=['subjects']).reset_index()
# pivot -> rows to columns (pivoting)
```

```python
pd.pivot_table(melt_df,index=['name','stream'],columns=['subjects'],values='marks',aggfunc='sum',margins=True,margins_name='Total')
```

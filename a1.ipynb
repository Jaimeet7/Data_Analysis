```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
week = ['w1','w2','w3','w4']

store_1_sales = [6000,8000,12000,14000]
store_2_sales = [96000,104000,98000,102000]
```

    Covarience = Standard deviation / mean *100

```python
fig,axis = plt.subplots(2,2,figsize=(5,5))
s1 = sns.scatterplot(x=week,y=store_1_sales,ax=axis[0][0])
s2 = sns.scatterplot(x=week,y=store_2_sales,ax=axis[1][0])
# s1.hlines(y=np.mean(store_1_sales),xmin='w1',xmax='w4',colors='pink')
# s2.hlines(y=np.mean(store_2_sales),xmin='w1',xmax='w4',colors='pink')
# s1.vlines(x=week,ymin=np.mean(store_1_sales),ymax=store_1_sales,colors='green')
# s2.vlines(x=week,ymin=np.mean(store_2_sales),ymax=store_2_sales,colors='green')
plt.tight_layout()
```

Co effitience of varaiance - Used to calculate wether variance is high or not

```python
s1 = np.mean(store_1_sales)
std1 = np.std(store_1_sales)
cv1 = (std1/s1)*100
print(f'store 1 has a weekly revenue of {s1} and most of the week the revenu falls between {s1} +- {std1.round(2)}')


print(f'store 1 has {cv1.round(2)} % variation from the {s1}')
```

```python
s2 = np.mean(store_2_sales)
std2 = np.std(store_2_sales)
cv2 = (std2/s2)*100
print(cv2)
print(f'store 2 has a weekly revenue of {s2} and most of the week the revenu falls between {s2} +- {std2.round(2)}')\

print(f'store 1 has {cv2.round(2)} % variation from the {s2}')
```
